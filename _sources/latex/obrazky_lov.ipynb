```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import solve_ivp
```

```python
parametry = [0.15, 0.3]
params = {'style':["-","--","-.",":",">-"], 'color':['gray','C0','C1','C2','C3']}
x = np.linspace(0,1,30)
df = pd.DataFrame(index = x)
df.index.name = "velikost populace"
r,K = 1,1
df["logistický růst"] = r*x*(1-x/K)
for h in parametry:
    df[f"lov intenzity {h}"] = 0*x + h
df.columns = ["logistický růst", "malý lov", "intenzivní lov"]

fig, ax = plt.subplots(figsize=(4,3))
df.plot(**params, ax = ax)
# plt.legend(loc='upper right')
ax.set(ylabel="rychlost růstu a poklesu")
# plt.grid()
plt.tight_layout()
plt.savefig("log_konst.pdf")
```
